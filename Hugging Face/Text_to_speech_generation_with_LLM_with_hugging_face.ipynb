```python
from transformers import pipeline
text = "Welcome to our platform. We're excited to have you here and can’t wait to help you get started. Whether you're exploring new features or building something amazing, everything you need is right at your fingertips. Let’s begin this journey together and make something great."
pipe = pipeline("text-to-speech", model="suno/bark-small", device = "cuda" )
output = pipe(text)
```

```python
output
```

```python
from  IPython.display import Audio
Audio(output["audio"], rate = output["sampling_rate"])
```
